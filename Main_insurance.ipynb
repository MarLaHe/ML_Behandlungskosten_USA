## 1. Geschäftsmodell verstehen

Das amerikanische Gesundheitssystem kennt keine allgemein verpflichtende Krankenversicherung für alle Staatsbürger, wie es im deutschen System üblich ist. Vielmehr ist eine Versicherung für die meisten Amerikaner optional. Es hat aber in den letzten Jahren einen großen Zuwachs an Versicherten gegeben - so waren im Jahr 2022 92% der Amerikaner versichert, davon 65.6% privat und 36.1% gesetzlich (Keisler-Starkey, Bunch & Lindstrom, 2023). Die gesetzlichen Versicherungen sind für drei ausgewählte Personenkreise verfügbar und werden vom Staat finanziert:
* Medicare für ältere Menschen ab 65 Jahren, Menschen mit anerkannten Behinderungen oder akutem Nierenversagen 
* Medicaid und CHIP für alle Kinder sowie Erwachsene, deren Einkommen unter der Armutsgrenze liegt
* Tricare für aktive und ehemalige Streitkräfte des US-Militärs und deren Angehörige
<br>(USAGov, 2024).

<br> Der Großteil der US-Bürger fällt nicht in die Zuständigkeit der staatlichen Versicherungsprogramme. Sie haben die Möglichkeit, sich eigenständig oder gebunden an ihren Arbeitgeber privat zu versichern. Wobei die beliebteste Option die eben genannten arbeitgebergebundenen Gruppen-Versicherungen sind, diese decken 54.5% der Beölkerung ab (Keisler-Starkey, Bunch & Lindstrom, 2023). Hier gibt es unterschiedliche Konzepte von anteiliger Übernahme der Prämien durch die Unternehmen bis hin zu vollständiger Kostenübernahme durch die Mitarbeiter. Der große Nachteil der an Unternehmen gebundenen Krankenversicherungen ist, dass der Krankenversicherungsanspruch oft verfällt, sobald die Versicherten nicht mehr im Unternehmen beschäftigt sind. 
<br> Der Markt der privaten Krankenversicherungen ist breit gefächert, es gibt an die Tausend verschiedene Versicherer und es werden vielfältige Leistungsspektren zu unterschiedlichsten Preisen angeboten. Die monatlichen Kosten sind lt. Masterson & Megna (2024) meist abhängig vom Alter, dem Wohnort, dem Raucherstatus, der Höhe des Selbstbehalts (es gibt unterschiedliche Versicherungspläne mit gestaffelter Höhe des Eigenanteils), dem Einkommen, der Art der Versicherungsleistung und der Anzahl der (mit)versicherten Personen. Für einzelne junge Versicherte von 21 Jahren liegen die Prämien demnach bei etwa 400 US-Dollar (etwa 367 €) pro Monat.

Wir bewegen uns für die Studienarbeit auf dem Markt der privaten Krankenversicherungen der USA. Für eine solche soll analysiert werden, ob und wenn ja, welche Merkmale und Kombinationen von individuell ausgeprägten Merkmalen die Höhe der anfallenden Ansprüche bzw. Behandlungskosten von Versicherten beeinflussen. Anhand der verschiedenen Merkmale soll im Idealfall ein Modell geschaffen werden, welches für potentielle Neukunden den zu erwartenden Versicherungsanspruch/die Behandlungskosten schätzt. Anhand dieser Schätzung könnten dann in einer Weiterentwicklung des Projektes zum Beispiel individuelle monatliche Versicherungsprämien (Betrag, den der Versicherungsnehmer für den in einem Versicherungsvertrag vom Versicherer festgelegten Versicherungsschutz zu zahlen hat) für Neukunden festgelegt werden.

### Auswahl der Data Mining-Methoden

Wie bereits zu Beginn schon angedeutet, bestätigt sich auch nach der explorativen Datenanalyse, dass das beschriebene Problem mit Machine Learning Methoden aus dem Bereich des überwachten Lernens gelöst werden sollte, da wir einen Datensatz haben, der die zu erklärende Variable bzw. gewünschten Lösungen oder "Labels" bereits enthält. Im überwachten Lernen wird vereinfacht beschrieben eine Funktion f(x) erlernt, die jedem Input x einen Output y zuordnet. Das Ziel ist es, die Funktion so genau wie möglich zu bestimmen, um anhand dieser möglichst präzise Vorhersagen für neue Inputs treffen zu können.
<br>Innerhalb des überwachten Lernens ist für das beschriebene Problem die Anwendung der Regression sinnvoll, da die numerische Zielgröße "claim" erklärt und vorhergesagt werden soll.
<br>Schwierigkeiten könnten auftreten in der Anpassung des Modells, da als erklärende Variablen sowohl numerische als auch kategoriale vorliegen.

Es kommen für das beschriebene Geschäftsproblem verschiedene Methoden zur Lösung infrage:

* polynomiale Regression:Drückt die Beziehung zwischen abhängigen und unabhängigen Variablen durch eine Polynomgleichung aus, es können dadurch komplexere Beziehungen zwischen den Variablen modelliert werden, die nicht durch eine einfache lineare Beziehung beschrieben werden können. Es können auch Vorhersagen getroffen werden.
* Entscheidungsbaum-Regression: Erzeugt eine Baumstruktur und macht Vorhersagen durch den Vergleich von Merkmalen im Datensatz. Ein DecisionTree kann schnell anfällig für Overfitting werden.
* Random-Forest-Regression: Ermöglicht genauere Vorhersagen durch die Kombination mehrerer Entscheidungsbäume. Ein Random Forest wird durch die Kombination vieler Bäume weniger anfällig für Overfitting.

<br> Die vorgeschlagenen Methoden werden angewendet und die Qualität und Leistung der Modelle durch Kreuzvalidierung und vergleichende Maße wie 
<br>r2 (Bestimmtheitsmaß, Bewertung der Güte des Regressionsmodells), 
<br>mse (mittlere quadratische Abweichung zwischen den tatsächlichen und den vorhergesagten Werten, Bewertung der Genauigkeit eines Modells im Vergleich), 
<br>rmse (Wurzel der mittleren quadratischen Abweichung = Größe des Fehlers, den das System im Mittel bei Vorhersagen macht - in der Einheit der untersuchten Variablen!) bewertet.

Der Random Forest erreicht auf die Testdaten mit 0.79 das höchste Bestimmtheitsmaß, zudem mit einer geringen Bandbreite an Werten in den Keuzvalidierungen.
```python
#Merkmale auswählen
columns_to_drop = ['PatientID', 'claim', 'claim_category']
X=df_model.drop(columns_to_drop, axis=1)
y=df_model['claim']
```

```python
#X_train um Polynom erweitern
X_poly_train=poly_features.fit_transform(X_train)
```

```python
# Vorhersagen für Testdaten erstellen
y_pred_test = lin_reg_poly.predict(X_poly_test)
# Vorhersagen für die Trainingsdaten erstellen
y_pred_train = lin_reg_poly.predict(X_poly_train)
```

```python
#allgemeine Bewertung des Models
rmse = MSE(y_test, y_pred_test)**(1/2)
r2 = R2(y_test, y_pred_test)
print("rmse:", rmse.round(2))
print("r2", r2.round(2))
```

Mit einem R^2 von 0.78 auf die Trainingsdaten scheint das Modell zunächst gut abzuschneiden - es wird demnach 78% der Varianz der abhängigen Variablen erklärt. Der recht gute Wert kann aber auch auf Overfitting hindeuten. Um auf Overfitting zu untersuchen, werden nun noch die MSE-Werte der Trainings-, Test- und Kreuzvalidierungsdatensätze verglichen:

```python
#MSE auf die verschiedenen Datensätze vergleichen, um Over/Underfitting zu bewerten
print ('Train MSE: {: .2f}'.format(MSE(y_train, y_pred_train)))
print ('Test MSE: {: .2f}'.format(MSE(y_test, y_pred_test)))

cv_mse=-cross_val_score(lin_reg_poly, X_train, y_train, scoring="neg_mean_squared_error", cv=4)
print ('CV MSE: {: .2f}'.format(cv_mse.mean()))
```

Der CV MSE ist deutlich größer, als der Train MSE. Dies kann auf Overfitting der Schätzfunktion auf das Trainingsset hindeuten - das heißt, das Modell ist zu stark an die Trainingsdaten angepasst und nicht in der Lage, gut auf neuen, unbekannten Daten zu generalisieren. Mit einer höhergradigen Polynomfunktion wird dies sogar noch ausgeprägter (getestet und hier dann wieder weg gelassen).

Der Test MSE liegt zwischen dem Train und CV MSE. Dies deutet darauf hin, dass das Modell weder zu gut noch zu schlecht auf den neuen Daten abschneidet. Das Modell ist also in der Lage, zumindest einige Muster in den Testdaten zu erfassen, ohne sie zu stark anzupassen.

Es könnte nun die Modellkomplexität reduziert werden z.B. durch Feature-Auswahl oder Regularisierung mit der Lasso-Regression oder Elastic-Net-Regression oder theoretisch auch durch die Anwendung einer niedrigeren Polynomgleichung - wir haben hier aber schon nur ein Polynom zweiten Grades.

Mit 81% ist R^2 für das erste Modell recht hoch, der RMSE ist etwas höher als beim besten Tree. Womöglich haben wir es hier zunächst wieder mit Overfitting zu tun.

```python
#MSE auf die verschiedenen Datensätze vergleichen, um Over/Underfitting zu bewerten
print ('Train MSE: {: .2f}'.format(MSE(y_train, y_pred_train)))
print ('Test MSE: {: .2f}'.format(MSE(y_test, y_pred_test)))

cv_mse=-cross_val_score(rf, X_train, y_train, scoring="neg_mean_squared_error", cv=4)
print ('CV MSE: {: .2f}'.format(cv_mse.mean()))
```

Der Train MSE ist am höchsten, was bedeutet, dass das Modell am schlechtesten auf die Trainingsdaten passt. Dies kann darauf hindeuten, dass das Modell zu einfach ist bzw. die Anpassung an die zugrunde liegenden Muster der Daten gering ist oder auch dass Rauschen in den Daten vorliegt.
<br>Der im Vergleich recht niedrige Test MSE bedeutet, dass das Modell auf den Testdaten besser abschneidet als auf den Trainingsdaten, was ungewöhnlich ist und auf zufällige Variationen in den Testdaten oder eine bestimmte Struktur in den Testdaten hinweisen kann, die das Modell besser erfasst als für die Trainingsdaten.
<br>Der CV MSE, der etwa in der Mitte zwischen Train und Test MSE liegt, deutet darauf hin, dass das Modell auf neuen, nicht gesehenen Daten mittelmäßig abschneidet. Dies ist ein Hinweis darauf, dass das Modell nicht optimal generalisiert und anfällig für Overfitting sein könnte.
